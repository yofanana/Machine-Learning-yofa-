```python
#create if not exist
file = open("aa.txt","w")  
for a in range(0,100):
    file.write(str(a)+'\n')
```

```python
file.close()
f = open("aa.txt","r+")
print (f.name)
```

```python
#get line count
count = len(f.readlines())
print ("count line: %s" % count)
f.seek(0,0)
```

```python
a=0
while a<count:
    print ("line %s : %s" % (a,str(f.readline())))
    a=a+1
f.close()
```

```python
#其他实现方式
#打开文件的最好方式（练习关闭文件除外）
with open("aa.txt", 'r') as fp:
    for (l,v) in enumerate(fp):
        print("行数 %d, 值 %s"%(l+1,v), end="\n")
        
```
